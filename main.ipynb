# Importing packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, roc_auc_score
import pickle
```

# Data Analysis Phase

```python
dataset = pd.read_csv("kaggle_diabetes.csv")
print(dataset.shape)
dataset.head()
```

```python
dataset.describe()
```

```python
target_column = "Outcome"
numerical_column = dataset.columns.drop(target_column)
output_rows = dataset[target_column]
dataset.drop(target_column,axis=1,inplace=True)
```

```python
scaler = MinMaxScaler()
scaler.fit(dataset)
t_df = scaler.transform(dataset)
```

```python
t_df
```

# Splitting Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(t_df, output_rows, test_size=0.25, random_state=0)
```

```python
print('X_train:',np.shape(X_train))
print('y_train:',np.shape(y_train))
print('X_test:',np.shape(X_test))
print('y_test:',np.shape(y_test))
```

# Testing Models

```python
from sklearn.naive_bayes import GaussianNB
gaussian_naive_bayes_classifier = GaussianNB()
gaussian_naive_bayes_classifier.fit(X_train, y_train)
filename = "models/gaussian-naive-bayes-classifier.pkl"
pickle.dump(gaussian_naive_bayes_classifier, open(filename, "wb"))
loss_and_metrics = gaussian_naive_bayes_classifier.score(X_test, y_test)
y_pred = gaussian_naive_bayes_classifier.predict(X_test)
print("F1-Score: ", f1_score(y_test, y_pred))
print("Accuracy: ", round(loss_and_metrics*100, 2), "%")
```

```python
from sklearn import svm
support_vector_classifier = svm.SVC()
support_vector_classifier.fit(X_train, y_train)
filename = "models/support-vector-classifier.pkl"
pickle.dump(support_vector_classifier, open(filename, "wb"))
loss_and_metrics = support_vector_classifier.score(X_test, y_test)
y_pred = support_vector_classifier.predict(X_test)
print("F1-Score: ", f1_score(y_test, y_pred))
print("Accuracy: ", round(loss_and_metrics*100, 2), "%")
```

```python
from sklearn.linear_model import LogisticRegression
logistic_regression_classifier = LogisticRegression()
logistic_regression_classifier.fit(X_train, y_train)
filename = "models/logistic-regression-classifier.pkl"
pickle.dump(logistic_regression_classifier, open(filename, "wb"))
loss_and_metrics = logistic_regression_classifier.score(X_test, y_test)
y_pred = logistic_regression_classifier.predict(X_test)
print("F1-Score: ", f1_score(y_test, y_pred))
print("Accuracy: ", round(loss_and_metrics*100, 2), "%")
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_classifier = RandomForestClassifier(n_estimators=100)
random_forest_classifier.fit(X_train, y_train)
filename = 'models/random-forest-classifier.pkl'
pickle.dump(random_forest_classifier, open(filename, 'wb'))
loss_and_metrics = random_forest_classifier.score(X_test, y_test)
y_pred = random_forest_classifier.predict(X_test)
print("F1-Score: ", f1_score(y_test, y_pred))
print("Accuracy: ", round(loss_and_metrics*100, 2), "%")
```

```python
from sklearn.ensemble import AdaBoostClassifier
adaBoost_classifier = AdaBoostClassifier(n_estimators=50)
adaBoost_classifier.fit(X_train, y_train)
filename = 'models/adaboost-classifier.pkl'
pickle.dump(adaBoost_classifier, open(filename, 'wb'))
loss_and_metrics = adaBoost_classifier.score(X_test, y_test)
y_pred = adaBoost_classifier.predict(X_test)
print("F1-Score: ", f1_score(y_test, y_pred))
print("Accuracy: ", round(loss_and_metrics*100, 2), "%")
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gradient_boosting_classifier = GradientBoostingClassifier(n_estimators=100)
gradient_boosting_classifier.fit(X_train, y_train)
filename = 'models/gradient-boosting-classifier.pkl'
pickle.dump(gradient_boosting_classifier, open(filename, 'wb'))
loss_and_metrics = gradient_boosting_classifier.score(X_test, y_test)
y_pred = gradient_boosting_classifier.predict(X_test)
print("F1-Score: ", f1_score(y_test, y_pred))
print("Accuracy: ", round(loss_and_metrics*100, 2), "%")
```
